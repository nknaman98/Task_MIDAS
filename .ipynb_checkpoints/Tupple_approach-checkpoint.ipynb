## dropping out duplicate entries

    

```python
df1 = df1.drop_duplicates()
```

```python
df1
```

```python
processed= df1.dropna()
```

```python
processed
```

```python
for index,row in processed.iterrows():
    val=row['product_category_tree']
    if(val.find('>>')!=-1):
        
        parts = val.split(">>")
        category_head=parts[0][2:]
        if(category_head[-1]==']'):
            category_head = category_head[:-2]
        val=category_head
    else:
        val= "NA"
```

```python
processed
```

```python
processed = processed[processed.product_category_tree != 'N/A']
```

```python
processed
```

```python
def cat(s):
    if(s.find('>>')!=-1):  
        parts = s.split(">>")
        category_head=parts[0][2:]
        if(category_head[-1]==']'):
            category_head = category_head[:-2]
        return category_head
    else:
        return 'N/A'
```

```python
processed['product_category_tree']=processed['product_category_tree'].apply(cat)
```

```python
processed.groupby('product_category_tree').describe()
```

```python
processed
```

```python
def delete(x):
    del_list=['automation & robotics ','eyewear ','food & nutrition ','gaming ','health & personal care appliances ','home & kitchen ','home entertainment ','household supplies ','pet supplies ','sunglasses ','wearable smart devices ','ebooks ']
    if x in del_list:
        return False
    else:
        return True
```

```python
processed['bool']=processed['product_category_tree'].apply(delete)
```

```python

df2=processed[processed['bool']==True]
```

```python
df2
```

```python
plt.xticks(rotation=90)
plt.rcParams['figure.figsize']=(30,20)
sns.countplot(x='product_category_tree',data= df2)
```
